```python
import matplotlib.pyplot as plt

# Count customers who churned and who didn't
churn_counts = df["Churn"].value_counts()

# Create the bar chart
plt.figure(figsize=(6,4))
plt.bar(churn_counts.index, churn_counts.values)

# Add labels
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

# Show the plot
plt.show()
```

```python
df["Churn"].value_counts(normalize=True) * 100
```

```python
import matplotlib.pyplot as plt

churn_counts = df["Churn"].value_counts()

plt.figure(figsize=(6,4))
bars = plt.bar(churn_counts.index, churn_counts.values)

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + 50,
        int(height),
        ha='center'
    )

plt.show()
```

```python
gender_churn = pd.crosstab(df["gender"], df["Churn"])

gender_churn.plot(
    kind="bar",
    figsize=(6,4)
)

plt.title("Gender vs Churn")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)

plt.show()
```

```python
pd.crosstab(
    df["gender"],
    df["Churn"],
    normalize="index"
) * 100
```

```python
contract_churn = pd.crosstab(df["Contract"], df["Churn"])

contract_churn.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Contract Type vs Churn")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)

plt.show()
```

```python
pd.crosstab(
    df["Contract"],
    df["Churn"],
    normalize="index"
) * 100
```

```python
# Monthly Charges by Churn
plt.figure(figsize=(8,5))

plt.hist(df[df["Churn"] == "No"]["MonthlyCharges"], bins=30, alpha=0.6, label="No")
plt.hist(df[df["Churn"] == "Yes"]["MonthlyCharges"], bins=30, alpha=0.6, label="Yes")

plt.title("Monthly Charges Distribution by Churn")
plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6,5))

df.boxplot(column="MonthlyCharges", by="Churn")

plt.title("Monthly Charges by Churn")
plt.suptitle("")  # Removes the default pandas title
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df[df["Churn"] == "No"]["tenure"], bins=30, alpha=0.6, label="No")
plt.hist(df[df["Churn"] == "Yes"]["tenure"], bins=30, alpha=0.6, label="Yes")

plt.title("Tenure Distribution by Churn")
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")
plt.legend()

plt.show()
```
